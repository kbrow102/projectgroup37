```python
import requests
import json
import matplotlib
import matplotlib.pyplot as plt
import plotly 
import plotly.plotly as py     
import plotly.graph_objs as go  
import cufflinks as cf
import pandas as pd
```

```python
print("Block Party")

currencies = []

while True:
    coin = input(str("Enter currency to be added to portfolio: ")).lower()
    if coin == 'bitcoin':
        bitcoin_url = 'https://api.coinmarketcap.com/v1/ticker/bitcoin/'
        bitcoin_request = requests.get(bitcoin_url)
        bitcoin_reply = bitcoin_request.json()
        bitcoin_json = bitcoin_reply[0]
        currencies.append(bitcoin_json)
    elif coin == 'dogecoin':
        dogecoin_url = 'https://api.coinmarketcap.com/v1/ticker/dogecoin/'
        dogecoin_request = requests.get(dogecoin_url)
        dogecoin_reply = dogecoin_request.json()
        dogecoin_json = dogecoin_reply[0]
        currencies.append(dogecoin_json)
    elif coin == 'litecoin':
        litecoin_url = 'https://api.coinmarketcap.com/v1/ticker/litecoin/'
        litecoin_request = requests.get(litecoin_url)
        litecoin_reply = litecoin_request.json()
        litecoin_json = litecoin_reply[0]
        currencies.append(litecoin_json)
    elif coin == 'ripple':
        ripple_url = 'https://api.coinmarketcap.com/v1/ticker/ripple/'
        ripple_request = requests.get(ripple_url)
        ripple_reply = ripple_request.json()
        ripple_json = ripple_reply[0]
        currencies.append(ripple_json)
    elif coin == 'iota':
        iota_url = 'https://api.coinmarketcap.com/v1/ticker/iota/'
        iota_request = requests.get(iota_url)
        iota_reply = iota_request.json()
        iota_json = iota_reply[0]
        currencies.append(iota_json)
    elif coin == 'ethereum':
        ethereum_url = 'https://api.coinmarketcap.com/v1/ticker/ethereum/'
        ethereum_request = requests.get(ethereum_url)
        ethereum_reply = ethereum_request.json()
        ethereum_json = ethereum_reply[0]
        currencies.append(ethereum_json)
    elif coin == 'stop':
        break
    else:
        print("Sorry, that is not a recognized input")
        
currencydf = pd.DataFrame(currencies).set_index(['name', 'symbol'])

currencydf.iplot(kind = 'scatter', x = 'Year', y = 'Population', title='Syracuse Population',xTitle='Year',yTitle='Population')
```
