```python
import requests
import json
import matplotlib
import matplotlib.pyplot as plt
import plotly 
import plotly.plotly as py     
import plotly.graph_objs as go  
import cufflinks as cf
import pandas as pd
```

```python
print("Block Party\n")

print("Supported Currencies:\n-Bitcoin\n-Dogecoin\n-Litecoin\n-Ripple\n-Iota\n-Ethereum\n")

currencies = []
blockchain_list = []

print("Blockchain Analysis")

blockchain_url = 'https://api.blockchain.info/stats'
blockchain_response = requests.get(blockchain_url)
blockchain_info = blockchain_response.json()
blockchain_list.append(blockchain_info)
blockchaindf= pd.DataFrame(blockchain_list)
blockchaindf.columns = ['Block Size', 'Difficulty', 'Est. BTC Sent', 'Est. Transaction Vol. (USD)', 'Hash Rate', 'Market Price (USD)', 'Miners Revenue (BTC)', 'Miners Revenue (USD)', 'Minutes Between Blocks', 'Blocks Mined', 'Blocks Total', 'BTC Mined', 'TX', 'Next Retarget', 'Timestamp', 'Total BTC Sent', 'Total Fees (BTC)', 'Blockchain Total', 'Trade Volume (BTC)', 'Trade Volume (USD)',]
display(blockchaindf)

while True:
    coin = input(str("Enter currency to be added to portfolio: ")).lower()
    if coin == 'bitcoin':
        bitcoin_url = 'https://api.coinmarketcap.com/v1/ticker/bitcoin/'
        bitcoin_request = requests.get(bitcoin_url)
        bitcoin_reply = bitcoin_request.json()
        bitcoin_json = bitcoin_reply[0]
        currencies.append(bitcoin_json)
    elif coin == 'dogecoin':
        dogecoin_url = 'https://api.coinmarketcap.com/v1/ticker/dogecoin/'
        dogecoin_request = requests.get(dogecoin_url)
        dogecoin_reply = dogecoin_request.json()
        dogecoin_json = dogecoin_reply[0]
        currencies.append(dogecoin_json)
    elif coin == 'litecoin':
        litecoin_url = 'https://api.coinmarketcap.com/v1/ticker/litecoin/'
        litecoin_request = requests.get(litecoin_url)
        litecoin_reply = litecoin_request.json()
        litecoin_json = litecoin_reply[0]
        currencies.append(litecoin_json)
    elif coin == 'ripple':
        ripple_url = 'https://api.coinmarketcap.com/v1/ticker/ripple/'
        ripple_request = requests.get(ripple_url)
        ripple_reply = ripple_request.json()
        ripple_json = ripple_reply[0]
        currencies.append(ripple_json)
    elif coin == 'iota':
        iota_url = 'https://api.coinmarketcap.com/v1/ticker/iota/'
        iota_request = requests.get(iota_url)
        iota_reply = iota_request.json()
        iota_json = iota_reply[0]
        currencies.append(iota_json)
    elif coin == 'ethereum':
        ethereum_url = 'https://api.coinmarketcap.com/v1/ticker/ethereum/'
        ethereum_request = requests.get(ethereum_url)
        ethereum_reply = ethereum_request.json()
        ethereum_json = ethereum_reply[0]
        currencies.append(ethereum_json)
    elif coin == 'stop':
        break
    else:
        print("Sorry, that is not a recognized input")
        
currencydf = pd.DataFrame(currencies).set_index(['name','symbol'])
currencydf.columns = ['24Hr Volume (USD)', 'Available Supply', 'ID', 'Last Updated', 'Market Cap (USD)', 'Max Supply', 'Percent Change 1Hr', 'Percent Change 24Hr', 'Percent Change 7 Day', 'Price in BTC', 'Price in USD', 'Rank', 'Total Supply']
display(currencydf)

volumegraph = (currencydf.iplot(kind = 'bar', x = 'ID', y = '24Hr Volume (USD)', title='24 Hour Currency Volumes (USD)',xTitle='Currency',yTitle='Volume (USD)'))
pctchange1hr = (currencydf.iplot(kind = 'bar', x = 'ID', y = 'Percent Change 1Hr', title='1 Hour Percent Change',xTitle='Currency',yTitle='Percent Change'))
pctchange24hr = (currencydf.iplot(kind = 'bar', x = 'ID', y = 'Percent Change 24Hr', title='24 Hour Percent Change',xTitle='Currency',yTitle='Percent Change'))
pctchange7day = (currencydf.iplot(kind = 'bar', x = 'ID', y = 'Percent Change 7 Day', title='7 Day Percent Change',xTitle='Currency',yTitle='Percent Change'))
display(volumegraph, pctchange1hr, pctchange24hr, pctchange7day)
```
